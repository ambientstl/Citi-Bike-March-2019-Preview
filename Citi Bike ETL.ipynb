```python
import pandas as pd

path = "Resources/201803-citibike-tripdata.csv"
#import march 2018 data
df2018 = pd.read_csv(path)
```

```python
df2018.head()
```

```python
df2018["starttime"][0].split(" ")[1]
```

```python
#isolate start and end dates/times
df2018["start date"] = [df2018["starttime"][i].split(" ")[0] for i in range(len(df2018))]
```

```python
df2018["stop date"] = [df2018["stoptime"][i].split(" ")[0] for i in range(len(df2018))]
df2018["start time"] = [df2018["starttime"][i].split(" ")[1] for i in range(len(df2018))]
df2018["stop time"] = [df2018["stoptime"][i].split(" ")[1] for i in range(len(df2018))]
```

```python
#rename columns
df2018 = df2018.rename(columns={"tripduration": "trip duration", "bikeid": "bike id", "usertype": "user type"})
```

```python
columns = df2018.columns.tolist()
columns
```

```python
#reorder columns
columns = ['trip duration','start date','start time','stop date','stop time', 
         'start station id','start station name','start station latitude','start station longitude',
         'end station id','end station name','end station latitude','end station longitude',
         'bike id','user type','birth year','gender']

df2018 = df2018[columns]
df2018
```

```python
len(df2018)
```

```python
#some data cleaning, removing extra short or long trips, and probably falsely reported birth years
clean_df2018 = df2018.loc[(df2018["trip duration"] > 100) & (df2018["trip duration"] < 15000) & (df2018["birth year"] > 1930), :]
len(clean_df2018)
```

```python
#save cleaned data to csv
clean_df2018.to_csv("Resources/Clean_March_2018.csv")
```
